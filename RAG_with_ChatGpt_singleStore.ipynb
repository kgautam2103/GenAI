```python
# prompt: using langchain how to pass the query and the result to the open ai as prompt

from langchain.chat_models import ChatOpenAI
from langchain.prompts.chat import (
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate,
)

chat = ChatOpenAI(temperature=0)

template = """You are a helpful data analyst/assistant that can answer questions about the attendees of the US 21 days global meditation program.
            Use the context below to answer questions. \n\n Context: {context} \n\n Question: {question}"""
system_message_prompt = SystemMessagePromptTemplate.from_template(template)
human_template = "{question}"
human_message_prompt = HumanMessagePromptTemplate.from_template(human_template)
chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])

# This is your previous query
query = "tell me more about daniel?"

# Get relevant documents from your vector database
docs = docsearch.similarity_search(query)

# Extract the text content from the relevant documents
context = "\n\n".join([doc.page_content for doc in docs])

# Create the message to send to the chat model
message = chat_prompt.format_prompt(context=context, question=query).to_messages()

# Send the message to the chat model and get the response
response = chat(message)

print(response.content)
```
